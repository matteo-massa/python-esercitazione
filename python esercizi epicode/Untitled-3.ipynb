```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

# ==============================================================================
# 1. LE TUE CREDENZIALI
# ==============================================================================


CLIENT_ID = '930aa5e18e654975b111a4ce9639cc31'          # <-- SOSTITUISCI IL TUO CLIENT ID

CLIENT_SECRET = "5bdbd761b8c94a9a9e37fefe81d9a380"  # <-- SOSTITUISCI IL TUO CLIENT SECRET

REDIRECT_URI = 'http://127.0.0.1:8888'      # <-- USA QUELLO CHE HAI REGISTRATO

NOME_PLAYLIST = "playlist 2026"
DESCRIZIONE = "La mia collezione rock completa."
SCOPE = "playlist-modify-public"

# ==============================================================================
# 2. LISTA PRONTA (Nessun trattino da analizzare, è già codice Python)
# ==============================================================================

SONG_LIST = [
    ("Viva La Vida", "Coldplay"),
    ("Clocks", "Coldplay"),
    ("The Scientist", "Coldplay"),
    ("Yellow", "Coldplay"),
    ("Paradise", "Coldplay"),
    ("Fix You", "Coldplay"),
    ("Adventure of a Lifetime", "Coldplay"),
    ("Hymn for the Weekend", "Coldplay"),
    ("A Sky Full of Stars", "Coldplay"),
    ("Everglow", "Coldplay"),
    ("Starlight", "Muse"),
    ("Supermassive Black Hole", "Muse"),
    ("Time Is Running Out", "Muse"),
    ("Madness", "Muse"),
    ("Counting Stars", "OneRepublic"),
    ("Secrets", "OneRepublic"),
    ("Apologize", "OneRepublic"),
    ("Chasing Cars", "Snow Patrol"),
    ("Open Your Eyes", "Snow Patrol"),
    ("Run", "Snow Patrol"),
    ("Somewhere Only We Know", "Keane"),
    ("Everybody’s Changing", "Keane"),
    ("Use Somebody", "Kings of Leon"),
    ("Sex on Fire", "Kings of Leon"),
    ("Believe", "Imagine Dragons"),
    ("Demons", "Imagine Dragons"),
    ("It’s Time", "Imagine Dragons"),
    ("Shots", "Imagine Dragons"),
    ("Radioactive", "Imagine Dragons"),
    ("Pompeii", "Bastille"),
    ("Happier", "Bastille"),
    ("Riptide", "Vance Joy"),
    ("The A Team", "Ed Sheeran"),
    ("Hero", "Family of the Year"),
    ("Holes", "Passenger"),
    ("Waves", "Dean Lewis"),
    ("High Hopes", "Kodaline"),
    ("All I Want", "Kodaline"),
    ("Let Her Go", "Passenger"),
    ("Gale Song", "The Lumineers"),
    ("Do I Wanna Know?", "Arctic Monkeys"),
    ("R U Mine?", "Arctic Monkeys"),
    ("505", "Arctic Monkeys"),
    ("Arabella", "Arctic Monkeys"),
    ("Mr. Brightside", "The Killers"),
    ("When You Were Young", "The Killers"),
    ("Somebody Told Me", "The Killers"),
    ("The Way It Was", "The Killers"),
    ("The Pretender", "Foo Fighters"),
    ("Everlong", "Foo Fighters"),
    ("Best of You", "Foo Fighters"),
    ("Learn to Fly", "Foo Fighters"),
    ("Walk", "Foo Fighters"),
    ("Numb", "Linkin Park"),
    ("In the End", "Linkin Park"),
    ("Crawling", "Linkin Park"),
    ("Breaking the Habit", "Linkin Park"),
    ("New Divide", "Linkin Park"),
    ("Decode", "Paramore"),
    ("Still Into You", "Paramore"),
    ("Misery Business", "Paramore"),
    ("Ain’t It Fun", "Paramore"),
    ("The Kill", "Thirty Seconds to Mars"),
    ("Kings and Queens", "Thirty Seconds to Mars"),
    ("This Is War", "Thirty Seconds to Mars"),
    ("Closer", "Nine Inch Nails"),
    ("Head Like a Hole", "Nine Inch Nails"),
    ("Come a Little Closer", "Cage the Elephant"),
    ("Cigarette Daydreams", "Cage the Elephant"),
    ("Ain’t No Rest for the Wicked", "Cage the Elephant"),
    ("Little Lion Man", "Mumford & Sons"),
    ("I Will Wait", "Mumford & Sons"),
    ("Believer", "Imagine Dragons"),
    ("Warriors", "Imagine Dragons"),
    ("On Top of the World", "Imagine Dragons"),
    ("Youth", "Daughter"),
    ("Silhouettes", "Of Monsters and Men"),
    ("Little Talks", "Of Monsters and Men"),
    ("Mountain Sound", "Of Monsters and Men"),
    ("Holocene", "Bon Iver"),
    ("Skinny Love", "Bon Iver"),
    ("Yellow Flicker Beat", "Lorde"),
    ("Royals", "Lorde"),
    ("Team", "Lorde"),
    ("The Night We Met", "Lord Huron"),
    ("My Number", "Foals"),
    ("Spanish Sahara", "Foals"),
    ("Gooey", "Glass Animals"),
    ("Heat Waves", "Glass Animals"),
    ("Take Me Out", "Franz Ferdinand"),
    ("No You Girls", "Franz Ferdinand"),
    ("Reptilia", "The Strokes"),
    ("Last Nite", "The Strokes"),
    ("Someday", "The Strokes"),
    ("Under Cover of Darkness", "The Strokes"),
    ("Mountain at My Gates", "Foals"),
    ("High", "Sir Sly"),
    ("Electric", "Alina Baraz & Khalid"),
    ("Ready To Start", "Arcade Fire"),
    ("Wake Up", "Arcade Fire"),
    ("Papillon", "Editors"),
    ("Munich", "Editors"),
    ("An End Has a Start", "Editors"),
    ("The Funeral", "Band of Horses"),
    ("Casual Party", "Band of Horses"),
    ("Sweet Disposition", "The Temper Trap"),
    ("Fader", "The Temper Trap"),
    ("Somebody Else", "The 1975"),
    ("The Sound", "The 1975"),
    ("Robbers", "The 1975"),
    ("Sex", "The 1975"),
    ("Intro", "The xx"),
    ("Crystalised", "The xx"),
    ("Angels", "The xx"),
    ("On Hold", "The xx"),
    ("Breezeblocks", "alt-J"),
    ("Taro", "alt-J"),
    ("Tessellate", "alt-J"),
    ("Left Hand Free", "alt-J"),
    ("Home", "Edward Sharpe & The Magnetic Zeros"),
    ("Sedona", "Houndmouth"),
    ("Ophelia", "The Lumineers"),
    ("Sleep on the Floor", "The Lumineers"),
    ("Cleopatra", "The Lumineers"),
    ("Myth", "Beach House"),
    ("Space Song", "Beach House"),
    ("Lazuli", "Beach House"),
    ("Chinatown", "Bleachers"),
    ("Rollercoaster", "Bleachers"),
    ("Dirty Paws", "Of Monsters and Men"),
    ("Wait", "M83"),
    ("Midnight City", "M83"),
    ("Holes", "Mercury Rev"),
    ("Northern Lights", "Death Cab for Cutie"),
    ("I Will Follow You into the Dark", "Death Cab for Cutie"),
    ("Soul Meets Body", "Death Cab for Cutie"),
    ("Fire", "Barns Courtney"),
    ("Glitter & Gold", "Barns Courtney"),
    ("Kids", "MGMT"),
    ("Time to Pretend", "MGMT"),
    ("Electric Feel", "MGMT"),
    ("The Less I Know the Better", "Tame Impala"),
    ("Let It Happen", "Tame Impala"),
    ("Elephant", "Tame Impala"),
    ("Breathe", "The Prodigy"),
    ("Feel Good Inc.", "Gorillaz"),
    ("On Melancholy Hill", "Gorillaz"),
    ("19-2000", "Gorillaz"),
    ("Stylo", "Gorillaz"),
    ("Hey Now", "London Grammar"),
    ("Strong", "London Grammar"),
    ("Wasting My Young Years", "London Grammar"),
    ("Youth", "Daughter"),
    ("Amsterdam", "Nothing But Thieves"),
    ("Trip Switch", "Nothing But Thieves"),
    ("Sorry", "Nothing But Thieves"),
    ("Wolves", "Rag’n’Bone Man"),
    ("Human", "Rag’n’Bone Man"),
    ("Bohemian Rhapsody", "Queen"),
    ("Don’t Stop Me Now", "Queen"),
    ("Another One Bites the Dust", "Queen"),
    ("We Will Rock You", "Queen"),
    ("Under Pressure", "Queen & David Bowie"),
    ("Back in Black", "AC/DC"),
    ("Highway to Hell", "AC/DC"),
    ("TNT", "AC/DC"),
    ("Thunderstruck", "AC/DC"),
    ("Enter Sandman", "Metallica"),
    ("Nothing Else Matters", "Metallica"),
    ("Master of Puppets", "Metallica"),
    ("Immigrant Song", "Led Zeppelin"),
    ("Stairway to Heaven", "Led Zeppelin"),
    ("Whole Lotta Love", "Led Zeppelin"),
    ("Sweet Child O' Mine", "Guns N’ Roses"),
    ("Welcome to the Jungle", "Guns N’ Roses"),
    ("Paradise City", "Guns N’ Roses"),
    ("Smells Like Teen Spirit", "Nirvana"),
    ("Come as You Are", "Nirvana"),
    ("Lithium", "Nirvana"),
    ("Black Hole Sun", "Soundgarden"),
    ("Spoonman", "Soundgarden"),
    ("Jeremy", "Pearl Jam"),
    ("Alive", "Pearl Jam"),
    ("Even Flow", "Pearl Jam"),
    ("Californication", "Red Hot Chili Peppers"),
    ("Otherside", "Red Hot Chili Peppers"),
    ("Snow (Hey Oh)", "Red Hot Chili Peppers"),
    ("Dani California", "Red Hot Chili Peppers"),
    ("Under the Bridge", "Red Hot Chili Peppers"),
    ("Hotel California", "Eagles"),
    ("Take It Easy", "Eagles"),
    ("Fortunate Son", "CCR"),
    ("Proud Mary", "CCR"),
    ("Born to Be Wild", "Steppenwolf"),
    ("Smoke on the Water", "Deep Purple"),
    ("Hush", "Deep Purple"),
    ("Highway Star", "Deep Purple"),
    ("Paint It Black", "The Rolling Stones"),
    ("Satisfaction", "The Rolling Stones"),
    ("Start Me Up", "The Rolling Stones"),
    ("Sympathy for the Devil", "The Rolling Stones"),
    ("Gimme Shelter", "The Rolling Stones"),
    ("Hey Jude", "The Beatles"),
    ("Come Together", "The Beatles"),
    ("Here Comes the Sun", "The Beatles"),
    ("Let It Be", "The Beatles"),
    ("Heroes", "David Bowie"),
    ("Space Oddity", "David Bowie"),
    ("Starman", "David Bowie"),
    ("All Along the Watchtower", "Jimi Hendrix"),
    ("Purple Haze", "Jimi Hendrix"),
    ("Foxy Lady", "Jimi Hendrix"),
    ("Layla", "Eric Clapton"),
    ("Cocaine", "Eric Clapton"),
    ("Born in the U.S.A.", "Bruce Springsteen"),
    ("Dancing in the Dark", "Bruce Springsteen"),
    ("The River", "Bruce Springsteen"),
    ("We Are the Champions", "Queen"),
    ("Dream On", "Aerosmith"),
    ("Walk This Way", "Aerosmith"),
    ("Livin’ On a Prayer", "Bon Jovi"),
    ("Wanted Dead or Alive", "Bon Jovi"),
    ("Shot in the Dark", "Ozzy Osbourne"),
    ("Crazy Train", "Ozzy Osbourne"),
    ("Ace of Spades", "Motörhead"),
    ("Breaking the Law", "Judas Priest"),
    ("Barracuda", "Heart"),
    ("Chop Suey!", "System of a Down"),
    ("Toxicity", "System of a Down"),
    ("BYOB", "System of a Down"),
    ("Killing in the Name", "Rage Against the Machine"),
    ("Bulls on Parade", "Rage Against the Machine"),
    ("Bombtrack", "Rage Against the Machine"),
    ("Duality", "Slipknot"),
    ("Psychosocial", "Slipknot"),
    ("Before I Forget", "Slipknot"),
    ("Paranoid", "Black Sabbath"),
    ("Iron Man", "Black Sabbath"),
    ("War Pigs", "Black Sabbath"),
    ("The Trooper", "Iron Maiden"),
    ("Fear of the Dark", "Iron Maiden"),
    ("Hallowed Be Thy Name", "Iron Maiden"),
    ("Painkiller", "Judas Priest"),
    ("Angel of Death", "Slayer"),
    ("Raining Blood", "Slayer"),
    ("Holy Wars", "Megadeth"),
    ("Hangar 18", "Megadeth"),
    ("Symphony of Destruction", "Megadeth"),
    ("Cowboys from Hell", "Pantera"),
    ("Walk", "Pantera"),
    ("Roots Bloody Roots", "Sepultura"),
    ("The Heretic Anthem", "Slipknot"),
    ("One", "Metallica"),
    ("For Whom the Bell Tolls", "Metallica"),
    ("Fade to Black", "Metallica"),
    ("Battery", "Metallica"),
    ("Down with the Sickness", "Disturbed"),
    ("Indestructible", "Disturbed"),
    ("Stricken", "Disturbed"),
    ("Last Resort", "Papa Roach"),
    ("Between Angels and Insects", "Papa Roach"),
    ("Bodies", "Drowning Pool"),
    ("Rollin’", "Limp Bizkit"),
    ("Break Stuff", "Limp Bizkit"),
    ("Nookie", "Limp Bizkit"),
    ("My Own Summer", "Deftones"),
    ("Change (In the House of Flies)", "Deftones"),
    ("Diamond Eyes", "Deftones"),
    ("Hail to the King", "Avenged Sevenfold"),
    ("Nightmare", "Avenged Sevenfold"),
    ("Bat Country", "Avenged Sevenfold"),
    ("The Stage", "Avenged Sevenfold"),
    ("Unholy Confessions", "Avenged Sevenfold"),
    ("The Diary of Jane", "Breaking Benjamin"),
    ("Blow Me Away", "Breaking Benjamin"),
    ("I Miss the Misery", "Halestorm"),
    ("Love Bites", "Halestorm"),
    ("Renegades", "X Ambassadors"),
    ("Unsteady", "X Ambassadors"),
    ("Heathens", "Twenty One Pilots"),
    ("Stressed Out", "Twenty One Pilots"),
    ("Ride", "Twenty One Pilots"),
    ("Numb Little Bug", "Em Beihold"),
    ("Love Me Again", "John Newman"),
    ("Shut Up and Dance", "Walk the Moon"),
    ("Anna Sun", "Walk the Moon"),
    ("Waves", "Mr Probz"),
    ("Yellow (Live)", "Coldplay"),
    ("Something Just Like This", "Chainsmokers & Coldplay"),
    ("Paradise (Live)", "Coldplay"),
    ("Stubborn Love", "The Lumineers"),
    ("Where’s My Love", "SYML"),
    ("Lost On You", "LP"),
    ("Warriors", "Imagine Dragons"),
    ("Birds", "Imagine Dragons"),
    ("I Bet My Life", "Imagine Dragons"),
    ("Forever", "Lewis Capaldi"),
    ("Wonderwall", "Oasis"),
    ("Don’t Look Back in Anger", "Oasis"),
    ("Champagne Supernova", "Oasis"),
    ("Live Forever", "Oasis"),
    ("Supersonic", "Oasis"),
    ("Stand by Me", "Oasis"),
    ("Slide Away", "Oasis"),
    ("Cigarettes & Alcohol", "Oasis"),
    ("Some Might Say", "Oasis"),
    ("Morning Glory", "Oasis"),
    ("Stop Crying Your Heart Out", "Oasis"),
    ("The Masterplan", "Oasis"),
    ("Iris", "Goo Goo Dolls"),
    ("Slide", "Goo Goo Dolls"),
    ("Name", "Goo Goo Dolls"),
    ("Black Balloon", "Goo Goo Dolls"),
    ("Here Is Gone", "Goo Goo Dolls"),
    ("Sympathy", "Goo Goo Dolls"),
    ("Broadway", "Goo Goo Dolls"),
    ("Come to Me", "Goo Goo Dolls")
]

# ==============================================================================
# 3. MOTORE DELLO SCRIPT
# ==============================================================================

def get_spotify_instance():
    """Autenticazione con apertura browser automatica."""
    sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        redirect_uri=REDIRECT_URI,
        scope=SCOPE,
        show_dialog=True,
        open_browser=True
    ))
    return sp

def search_track_uri(sp, title, artist):
    """Cerca la canzone e restituisce l'URI."""
    query = f"track:{title} artist:{artist}"
    try:
        # Cerca track + artist
        results = sp.search(q=query, limit=1, type='track')
        if results['tracks']['items']:
            return results['tracks']['items'][0]['uri']
        
        # Se fallisce, prova solo con il titolo (utile se l'artista è un feat.)
        results = sp.search(q=f"track:{title}", limit=1, type='track')
        if results['tracks']['items']:
            return results['tracks']['items'][0]['uri']
            
        print(f"❌ Brano non trovato: {title} - {artist}")
        return None
    except Exception as e:
        print(f"Errore ricerca {title}: {e}")
        return None

if __name__ == '__main__':
    print(f"--- CREAZIONE {NOME_PLAYLIST} ---")
    
    # 1. Autenticazione
    sp = get_spotify_instance()
    user_id = sp.me()['id']
    print(f"Utente: {user_id}")

    # 2. Crea Playlist
    try:
        playlist = sp.user_playlist_create(
            user=user_id,
            name=NOME_PLAYLIST,
            public=True,
            description=DESCRIZIONE
        )
        playlist_id = playlist['id']
        print(f"Playlist creata! ID: {playlist_id}")
    except Exception as e:
        print(f"Errore creazione playlist: {e}")
        exit()

    # 3. Cerca Canzoni
    print(f"\nSto cercando {len(SONG_LIST)} canzoni...")
    track_uris = []
    
    for title, artist in SONG_LIST:
        uri = search_track_uri(sp, title, artist)
        if uri:
            track_uris.append(uri)
            
    # 4. Aggiungi Canzoni (in blocchi da 100)
    print(f"\nTrovati {len(track_uris)} brani. Aggiunta in corso...")
    
    for i in range(0, len(track_uris), 100):
        batch = track_uris[i:i + 100]
        sp.playlist_add_items(playlist_id, batch)
        print(f"-> Aggiunto blocco {i // 100 + 1}")

    print(f"\n FATTO! La '{NOME_PLAYLIST}' è pronta su Spotify.")
```
